```python
#ROCK PAPER SCISSORS GAME

#determing the choices
choices = ["rock", "paper", "scissors"]

#allowing user to enter their choice
user_choice = input("Pick your choice, enter in lower_case :")

#making a choice for the computer randomly

import random
choices = ["rock", "paper", "scissors"]
computer_choice = random.choice(choices)
print(computer_choice)

#determing the winner
if user_choice == computer_choice:
    print("It's a tie, try Again!")
elif (user_choice == "rock" and computer_choice == "scissors") or \
     (user_choice == "paper" and computer_choice == "rock") or \
     (user_choice == "scissors" and computer_choice == "paper"):
    print("You win, yaay!")
else:
    print("You lose!")
    
#diplaying both user choice and computer choice
print(f"Your choice : {user_choice}, computer choice : {computer_choice}.")

```

```python
#allowing user play multiple times by looping

import random
choices = ["rock", "paper", "scissors"]

while True:
    user_choice = input("Pick your choice, enter in lower_case : ") 
    computer_choice = random.choice(choices)
    print(f"Your choice: {user_choice}, Computer choice: {computer_choice}")
    
    if user_choice == computer_choice:
        print("It's a tie, try Again!")
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win, yaay!")
    else:
        print("You lose!") 
        
    play_again = input("Would you like to play again? yes/no:  ").lower()
    if play_again != "yes":
        print("Thank you for playing!")
        break
```

```python
#list of the options
choices = ["rock", "paper", "scissors"]

#getting the user's choice
user_choice = input("Pick either rock, paper or scissors, make sure it is in lowercase: ")

#generate a random choice for the computer
import random
choices = ["rock", "paper", "scissors"]
computer_choice = random.choice(choices)

#determining the winner of the game
if user_choice == computer_choice:
    print("A tie! Try Again")
elif (user_choice == "rock" and computer_choice == "scissors") or \
     (user_choice == "paper" and computer_choice == "rock") or \
     (user_choice == "scissors" and computer_choice == "paper"):
    print("You win!")
else:
    print("You Lose!")

#displaying both the user choices and computer choices
print(f"Your choice: {user_choice}, computer choice: {computer_choice}.")

#user playing multiple times
import random

choices = ["rock", "paper", "scissors"]

while True:
    user_choice = input("Pick either rock, paper or scissors, make sure it is in lowercase: ")

    computer_choice = random.choice(choices)

    print(f"Your choice : {user_choice} | Computer choice : {computer_choice} " )


    if user_choice == computer_choice:
        print("A tie! Try Again")
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "paper" and computer_choice == "rock") or \
         (user_choice == "scissors" and computer_choice == "paper"):
        print("You win!")
    else:
        print("You Lose!")
        
    #allowing user try again after entering wrong yes/no input.   
    while True:
        play_again = input("Do you want to play again? (yes/no): ").lower()
        if play_again == "yes":
            break  
        elif play_again == "no":
            print("Thanks for playing!")
            break  
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")
    
    if play_again == "no":
        break  
```
